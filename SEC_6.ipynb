```python
          # attack type of int
def strength(attack):
  if attack > 150:
    return 'strong'
  elif attack>100:
    return 'average'
  else:
    return 'weak'
```

```python
pokemon["strength"]  =  pokemon["Attack"].apply(strength)
pokemon
```

```python
#value_counts()  returns object(STRING) containing counts of unique values.   counts the valus of the given column
pokemon.value_counts(["strength"])

# HANDLE ALL THE CASES OTHERWISE IT WILL RETURN NaN
```

```python
del pokemon['strength']
```

```python
#  APPLY ALSO USE WITH MULTIPPLE COLUMNS
def score(row):
  attack= row['Attack']  # we can also use the index as a column
  defense= row["Defense"]
  speed= row["Speed"]
  return sum([attack,defense,speed])
```

```python
# pokemon['sum'] = pokemon["Attack"].apply(sum)
# pokemon
pokemon['TOTAL SCORE OF ATTACK,DEFENSE & SPEED'] = pokemon.apply(score ,axis = 1)
pokemon
```

```python

# we can use lamda
short_name = pokemon["Name"].apply (lambda x: x[0:5])
pokemon.insert(1 , 'shortnamess',short_name) # insert new column , name of column will be shortname, varible
pokemon
```
