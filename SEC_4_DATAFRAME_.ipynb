**dropna()**

```python

# remove missing values using dropna()
import pandas as pd
sep = pd.read_csv("/content/comma separated.csv", sep = ",", index_col=["index"])
sep
```

```python
# remove missing values using dropna()
sep.dropna() #remove every row where contains NaN
```

```python
sep.dropna(axis = 0)#remove all the row where contains NaN
```

```python
sep.dropna(axis = 1)#remove all the column where contains NaN
```

```python
# if whole row filled with missing values thats we check
# use how = "all"
sep.dropna(axis = 0 , how = "all") #remove row where all value is NaN   DEFAULT how="any"
```

```python
sep.dropna(axis = 1 , how = "all")
```

```python
sep.dropna(axis = 0 , how = "any", subset="hh")
sep.dropna(axis = 0 , how = "any", subset="gg")
sep.dropna(axis = 0 , how = "any", subset="hh")
```

**fillna()**

```python
import pandas as aa
dic = {
    "a":['1', '2', '3'],
    "b":[None, '4', None],
    "c":['5', '6', None],
      "d":['7', None, '8'],
    "c":['9', '10', None],
    }

df = aa.DataFrame(dic)
df
```

fills methods

1.   ffill and pad (DO THE EXACT SAME THING)uses previous valid value to fill the next none
2.   backfill and bfill (DO THE EXACT SAME THING)  use the last valid value to fill
the previous  NaN






```python
df.fillna(method="ffill") # both are same

df.fillna(method="pad")
```

```python
df.fillna(method="backfill") # both are same
df.fillna(method="bfill") # both are same
```
