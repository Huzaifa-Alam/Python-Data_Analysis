```python
import pandas as pd
pokemon = pd.read_csv('/content/Pokemon.csv', index_col='#')
pokemon
```

```python
# USING STRING IN PANDASS
# pokemon['Name'].upper() # does not work
pokemon['Name'].str.upper()
pokemon['Name'].str.lower()
pokemon['Name'].str.swapcase()

# REVERSE STRING                            (*!*)
pokemon['Name'].str[::-1]
  # USING .str()  WE CAN USE ANY STRING METHOD ON THIS


# pokemon['Attack']
# pokemon['Attack'].str.upper() # it does not work with non values though
```

```python
filter = pokemon["Name"].str.contains('q', case = True) # csae sensitive only small q will appear
filter = pokemon["Name"].str.contains('q', case = False)
pokemon[filter]
```

```python
# STARTSWITH  AND ENDSWITH
name_filter = pokemon["Type 1"].str.startswith("W") # remeber that its case sensitive
name_filter = pokemon["Name"].str.startswith("Do") # remeber that its case sensitive
pokemon[name_filter]
```

```python
end_filter = pokemon["Name"].str.endswith("er")
pokemon[end_filter]
```

```python
pokemon['Type 1'].str.replace('Grass' , 'HUZAIFA ' , regex = False)
```

```python
# REPLACE ()
# pokemon['Type 1'] =  pokemon['Type 1'].str.replace('Grass' , 'HUZAIFA ' , regex = False) # by default regex = False (REGEX MEANS + * /  -)

# pokemon[pokemon['Type 1']=='HUZAIFA']   # DISPLAY ALL THE UHZAIFA
pokemon['Type 1'] = pokemon['Type 1'].str.replace('Grass', 'HUZAIFA', regex=False)

# Displaying all rows where 'Type 1' is 'HUZAIFA'
pokemon[pokemon['Type 1'] == 'HUZAIFA']
```
