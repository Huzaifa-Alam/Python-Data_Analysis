```python

import pandas as pd
data = {'A': [1, 2, 3],
        'B': [4, 5, 61],
        'C': [7, 8, 91]}


data2 = {
    'A': [10, 11, 121],
    'B': [13, 14, 151],
    'C': [116, 17, 18],
    'D': [19, 20, 21]}

df1 = pd.DataFrame(data)
df2 = pd.DataFrame(data2)
```

```python

df1
```

```python
df2
```

```python
# concatenating (link (things) together in a chain or series) datasets
pd.concat([df1,df2] )
pd.concat([df1,df2] , axis = 1) #axis = o for columns
```

```python
# giving keys to our dataset
pd.concat([df1,df2] , keys=['Data  ONE', 'Data TWO'] )
pd.concat([df1,df2] , keys=['Data  ONE', 'Data TWO'] , axis = 1)
```

```python
# ignore the index
pd.concat([df1,df2], ignore_index=True)
```

```python
# use join (with original datasets)
data = {'A': [1, 2, 3],
        'B': [4, 5, 61],
        'C': [7, 8, 91]}


data2 = {
    'A': [10, 11, 121],
    'B': [4, 5, 6],
    'D': [116, 17, 18]} # D is different

df1 = pd.DataFrame(data)
df2 = pd.DataFrame(data2)
```

```python
pd.concat([df1,df2] , keys= ['1st DATA', '2nd DATA'] , axis = 1)
```

```python
# NOW USE JOIN
pd.concat([df1,df2]  ,  join = 'inner') #   WILL CONCATENATE COLUMNS THAT ARE SHARED BY BOTH DATASETS
# pd.concat([df1,df2]  ,  join = 'outer') #   by default COLUMNS THAT ARE SHARED BY all DATASETS
```

```python
# MERGE ( )  combine data sets
import pandas as pd
cols = [ 'Name','Salary']
company21 = pd.read_csv('/content/company21.csv' , sep = ';', usecols=cols)
company21
```

```python

# company22 = pd.read_csv('/content/company22.csv', sep = ';' , index_col='ID' , usecols=['ID', 'Age','Salary'])
company22 = pd.read_csv('/content/company22.csv', sep = ';', usecols=cols )
company22
```

```python
# OUTER JOIN ()  with marge

# marging
# outer will join everything
# apply merge method on pd in pandas like (pd.)
# how = 'inner' set as default
 # on =      in which columns do u wwant to apply
#  on = name of thecolumn we want to perform the join operation on*
#  indicator=True   False by default how its merge
# suffixes will be in tuple  means like this ()
pd.merge(left=company21, right =company22, how='outer',  on= "Name" ,suffixes=('_men' , "_women"), indicator=True )

```

```python
#how='inner'   contain in both data woh return karta hai
# /give me columns that are in both csv files  on name column
# if how is not defined it will consider all the columns/
pd.merge(left=company21, right =company22, how='inner',  on= "Name" ,suffixes=('_men' , "_women"), indicator=True )
```

```python
pd.merge(left = company22 , right = company21 , on = 'Name' , suffixes = ('1' , '2'))
# on = None  by default  if not define then its mean if both salary are same then
# it will appear like in ever below case  ( MARK case)
pd.merge(left = company22 , right = company21 , suffixes = ('1' , '2'))
```
