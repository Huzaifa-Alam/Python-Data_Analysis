VID (92) BARPLOTS

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cols = ["date", "low", "high" , "open", "close"]
apple_stock = pd.read_csv( "AAPL.csv" , usecols=cols , index_col="date" , parse_dates=["date"])
apple_stock.head(7)
```

```python
# resample method is often used with time-series data that has a datetime index.
years = apple_stock.resample("A").mean().round(2) # mean for every year # "A" yearly  # "AS" for year start
years.index = years.index.year
years
```

```python
years.plot(y = "close"  , kind = "bar") # giving mean of every year
# years.plot(y = "close"  , kind = "barh") # giving bar in horizontal
plt.ylabel("Mean")
plt.xlabel("Year")
```

```python
# years.plot(y= "close", kind = "bar" , color = "r")
years.plot(y= ["high" , "low"], kind = "bar" , color = ("yellow" , "blue"))# y is colunname here always be in list # coloyrs should be in tuple
```

VID(94) PIECHART

```python
# we are going to do which industry is most popular
richest = pd.read_csv("TopRichestInWorld.csv" )
richest
```

```python
industry_names = richest["Industry"].value_counts()
industry_names
```

```python
industry_names.index
```

```python
industry_names.values
```

```python
plt.pie(x=industry_names.values , labels=industry_names.index) # pie for piechart(circular diagram)
      # x = show the color inside the chart
plt.show() # hide all those unnecessary details
```

```python
labels = industry_names.index[:10]
labels
```

```python
values= industry_names.values[:10]
values
```

```python
plt.pie(x=values , labels=labels , autopct = "%.2f%%" ,colors= ("blue" , "orange" , "red" , "yellow" , "green" ))
# plt.pie(x=values , labels=labels , autopct = "%.2f%%" ) # if we wanna show in percentage we can use autopct
                                        ##%.2f two decimal places (can be 3or 4..)
plt.show()
```

VID(94) STYLES

```python
pokemon = pd.read_csv("Pokemon.csv" , index_col='#')
pokemon
```

```python
poke_groups = pokemon.groupby(by="Type 1")
max_attack = poke_groups["Attack"].max()
max_attack
```

```python
max_attack.plot(kind="bar" , color = ("red" , "blue"))
# which shows pshchich has most power of attack while poison has lowest
```

```python
# CHECK STYLES WHICH ARE AVALIBLE USE
plt.style.available
# using method plt.style.use("put any which are given below")
```

```python
plt.style.use('grayscale')
plt.style.use('dark_background')
max_attack.plot(kind="bar")
```

```python
def all_style():
  for styles in plt.style.available:
    plt.style.use(styles)
    max_attack.plot(kind= "bar" , figsize = (10,2)  )
    print(styles)
    plt.show() # wouldn't shot unnecessary things

all_style()
```
